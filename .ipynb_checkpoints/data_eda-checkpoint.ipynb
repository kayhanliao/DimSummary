```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
df = df.drop(labels = ['Unnamed: 0', 'Unnamed: 0.1'], axis = 1)
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
fig, ax = plt.subplots()
ax.hist(df['rating'])
ax.set_xlabel('Ratings')
ax.set_ylabel('Count')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(df.total_review)
ax.set_xlabel('Total Reviews')
ax.set_ylabel('Count')
plt.show()
```
